연습 문제

두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.

각각 5 x 5 이상의 크기를 가진다.
공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다.

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.arange(25).reshape(5,5), columns = ['a','b','c','d','e'])
df2 = pd.DataFrame(np.arange(25).reshape(5,5), columns = ['f','g','h','i','j'])

pd.merge(df1, df2, left_on = 'a', right_on = 'f')
```

```python
tips = pd.DataFrame({239: [29.03, 5.92, 'Male', "No", "Sat", "Dinner", 3],
                    240: [27.18, 2.00, "Female", "Yes", "Sat", "Dinner", 2],
                    241: [22.67, 2.00, "Male", "Yes", "Sat", "Dinner", 2],
                    242: [17.82, 1.75, "Male", "No", "Sat", "Dinner", 2],
                    243: [18.78, 3.00, "Female", "No", "Thur", "Dinner", 2]},
                   index = ["total_bill", "tip", "sex", "smoker", "day", "time", "size"])
tips = tips.T
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.tail()
```

```python
tips.describe()
```

```python
tips.groupby("sex").count()
```

```python
tips.groupby("sex").size()
```

```python
tips.groupby(["sex", "smoker"]).size()
```

```python
tips.pivot_table("tip_pct", "sex", "smoker", aggfunc = 'count', margins = True)
```

```python
tips.pivot_table("tip_pct", "sex", "smoker")
```

```python
tips.groupby(["sex", "smoker"])[["tip", "tip_pct"]].describe()
```

```python
def peak_to_peak(x):
    return x.max() - x.min()

tips.groupby(["sex", "smoker"])[["tip"]].agg(peak_to_peak)
```

```python
tips.groupby(["sex", "smoker"]).agg( peak_to_peak)[["total_bill"]]
```

```python
tips.groupby(["sex", "smoker"]).agg({'total_bill' : peak_to_peak})
```

```python
tips.pivot_table(index=['sex', 'smoker'])
```

```python
tips.pivot_table(['tip_pct', 'size'], index=['sex', 'day'], columns='smoker')
```

```python
tips.pivot_table('size', index=['time', 'sex', 'smoker'],
                 columns='day', aggfunc='sum', fill_value=0)
```
